```python
def print_board(board):
    for row in board:
        print(" ".join(str(num) if num != 0 else "." for num in row))


def is_valid(board, row, col, num):
    # Check if the number is not in the current row
    if num in board[row]:
        return False

    # Check if the number is not in the current column
    for i in range(9):
        if board[i][col] == num:
            return False

    # Check if the number is not in the 3x3 box
    box_row_start = (row // 3) * 3
    box_col_start = (col // 3) * 3
    for i in range(box_row_start, box_row_start + 3):
        for j in range(box_col_start, box_col_start + 3):
            if board[i][j] == num:
                return False

    return True


def find_empty_location(board):
    # Find an empty spot (denoted by 0) in the grid
    for i in range(9):
        for j in range(9):
            if board[i][j] == 0:
                return i, j
    return None


def solve_sudoku(board):
    empty_pos = find_empty_location(board)
    if not empty_pos:
        # No empty spot means the board is solved
        return True

    row, col = empty_pos

    for num in range(1, 10):  # Try numbers 1 to 9
        if is_valid(board, row, col, num):
            board[row][col] = num  # Place the number

            if solve_sudoku(board):
                return True  # If it solves, return True

            board[row][col] = 0  # Undo the move (backtrack)

    return False  # Trigger backtracking


# Example Sudoku board (0 represents empty cells)
board = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

print("Original Sudoku board:")
print_board(board)

if solve_sudoku(board):
    print("\nSolved Sudoku board:")
    print_board(board)
else:
    print("No solution exists.")
```

```python
import tkinter as tk
from tkinter import messagebox
import random

class SudokuSolverGUI:
    def __init__(self, root):
        self.root = root
        self.root.title("Sudoku Solver")

        self.board = [[0 for _ in range(9)] for _ in range(9)]
        self.entries = [[None for _ in range(9)] for _ in range(9)]

        self.create_grid()
        self.create_buttons()

        self.restart_game()

    def create_grid(self):
        for row in range(9):
            for col in range(9):
                entry = tk.Entry(self.root, width=3, font=('Arial', 18), justify='center')
                entry.grid(row=row, column=col, padx=5, pady=5)

                # Color 3x3 grid sections differently
                if (row // 3 + col // 3) % 2 == 0:
                    entry.config(bg='#D3D3D3')

                self.entries[row][col] = entry

    def create_buttons(self):
        solve_button = tk.Button(self.root, text="Solve", command=self.solve_puzzle, width=10, font=('Arial', 14))
        solve_button.grid(row=10, column=0, columnspan=3, pady=20)

        clear_button = tk.Button(self.root, text="Clear", command=self.clear_grid, width=10, font=('Arial', 14))
        clear_button.grid(row=10, column=3, columnspan=3, pady=20)

        restart_button = tk.Button(self.root, text="Restart", command=self.restart_game, width=10, font=('Arial', 14))
        restart_button.grid(row=10, column=6, columnspan=3, pady=20)

    def restart_game(self):
        # Generate a new random Sudoku puzzle
        self.board = self.generate_puzzle()
        self.display_board()

    def clear_grid(self):
        for row in range(9):
            for col in range(9):
                self.entries[row][col].delete(0, tk.END)
                self.entries[row][col].config(fg="black")

    def display_board(self):
        # Display the current board with pre-filled values
        self.clear_grid()
        for row in range(9):
            for col in range(9):
                if self.board[row][col] != 0:
                    self.entries[row][col].insert(0, str(self.board[row][col]))
                    self.entries[row][col].config(fg="blue")

    def get_board(self):
        for row in range(9):
            for col in range(9):
                val = self.entries[row][col].get()
                self.board[row][col] = int(val) if val.isdigit() else 0
        return self.board

    def is_valid(self, board, row, col, num):
        # Check if the number is not in the current row
        if num in board[row]:
            return False

        # Check if the number is not in the current column
        for i in range(9):
            if board[i][col] == num:
                return False

        # Check if the number is not in the 3x3 box
        box_row_start = (row // 3) * 3
        box_col_start = (col // 3) * 3
        for i in range(box_row_start, box_row_start + 3):
            for j in range(box_col_start, box_col_start + 3):
                if board[i][j] == num:
                    return False

        return True

    def find_empty_location(self, board):
        for i in range(9):
            for j in range(9):
                if board[i][j] == 0:
                    return i, j
        return None

    def solve_sudoku(self, board):
        empty_pos = self.find_empty_location(board)
        if not empty_pos:
            return True

        row, col = empty_pos

        for num in range(1, 10):
            if self.is_valid(board, row, col, num):
                board[row][col] = num

                if self.solve_sudoku(board):
                    return True

                board[row][col] = 0

        return False

    def solve_puzzle(self):
        self.get_board()

        if self.solve_sudoku(self.board):
            self.display_solution()
        else:
            messagebox.showerror("Error", "No solution exists!")

    def display_solution(self):
        for row in range(9):
            for col in range(9):
                if not self.entries[row][col].get().isdigit():
                    self.entries[row][col].insert(0, str(self.board[row][col]))
                    self.entries[row][col].config(fg="green")

    # Function to generate a valid Sudoku puzzle
    def generate_puzzle(self):
        # Step 1: Generate a fully solved valid board
        board = [[0 for _ in range(9)] for _ in range(9)]
        self.solve_sudoku(board)
        
        # Step 2: Remove random numbers to create the puzzle (adjust difficulty)
        self.remove_numbers(board)
        return board

    # Function to generate a complete, random Sudoku solution
    def solve_sudoku(self, board):
        empty_pos = self.find_empty_location(board)
        if not empty_pos:
            return True

        row, col = empty_pos

        nums = list(range(1, 10))
        random.shuffle(nums)  # Shuffle numbers to create randomness
        for num in nums:
            if self.is_valid(board, row, col, num):
                board[row][col] = num

                if self.solve_sudoku(board):
                    return True

                board[row][col] = 0

        return False

    # Remove random numbers from the solved board to create a playable puzzle
    def remove_numbers(self, board, num_holes=40):
        count = 0
        while count < num_holes:
            row = random.randint(0, 8)
            col = random.randint(0, 8)

            if board[row][col] != 0:
                board[row][col] = 0
                count += 1


# Main code to run the Sudoku Solver GUI
if __name__ == "__main__":
    root = tk.Tk()
    app = SudokuSolverGUI(root)
    root.mainloop()
```
